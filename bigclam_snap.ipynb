```python
import numpy as np
import networkx as nx
import operator
import matplotlib.pyplot as pl
import os
%matplotlib inline
```

```python
pl.close('all')
N=100
K=5
alpha=.03
delta=.01
F=np.random.dirichlet(alpha*np.ones(K),N)
#F=np.random.rand(N,K)
#    F=[]        
#    for i in range(N):
#        n_com=np.random.poisson(3)
#        com_idx=np.random.randint(0,K,n_com)
#        f=np.zeros(K)
#        f[com_idx]=1
#        f=f/sum(f)
#        F+=[f]
#        
#    F=np.array(F)

#F=np.random.binomial(1,.2,(N,K))
C=np.argmax(F,1)    
theta=np.dot(F,F.T)   
G=nx.Graph()    
G.add_nodes_from(range(N))    

for u in range(N):
    for v in range(u+1,N):
        p=np.random.poisson(theta[u,v]+delta**2)
        #p=np.random.poisson(delta**2)
        if p>0:
            G.add_edge(u,v)   
deg=[d[1] for d in G.degree_iter()] 
```

```python
pos=nx.spring_layout(G)
pl.figure(figsize=(6*6,6))
#cls=range(K)
cls={0:'r',1:'b',2:'g',3:'y',4:'m'}
node_color=[cls[k] for k in C] 
pl.subplot(1,K+1,0)
nx.draw_networkx(G,pos,node_color=node_color,cmap=pl.cm.jet,with_labels=False)
pl.axis('off')
K=F.shape[1]
for k in range(K):
    pl.subplot(1,K+1,k+1)    
    nx.draw_networkx_edges(G,pos,alpha=.2)
    #fig.set_facecolor("#000001")
    for u in pos:
        nx.draw_networkx_nodes(G,nodelist=[u],pos={u:pos[u]},alpha=.2+.8*F[u,k],node_color=cls[k],with_labels=False)
    pl.axis('off')
pl.savefig('networks.pdf',bbox_inches='tight')
```

```python
def write_snap_file(G,fname):
    with open(fname,'wb') as file:
        file.writelines(['# Undirected graph (each unordered pair of nodes is saved once): sample_network.txt \n','# Nodes: '+str(G.number_of_nodes())+' Edges: '+str(G.number_of_edges())+'\n','# NodeId\tNodeId\n'])                              
        nx.write_edgelist(G,file,data=False,delimiter='\t')

def run_bigclam_snap(infile,outfile):
    os.system(" /Users/kirill.trapeznikov/Downloads/Snap-2.2/examples/bigclam/bigclam -c:5 -i:'/Users/kirill.trapeznikov/Documents/community/"+infile+"' -o:'/Users/kirill.trapeznikov/Documents/community/"+outfile+ "' ")
```

```python
fname='sample_network.txt'
fout='sample_network_com.txt'
write_snap_file(G,fname)
run_bigclam_snap(fname,fout)
```

```python
with open(fout+'cmtyvv.txt') as file:
    pred_coms=[]
    C=[]
    kidx=[]
    for line in file:
        c=np.zeros(len(G))
        nodes=np.array([int(x) for x in line.split()])
        c[nodes]=1.
        pred_coms+=[nodes]
        C+=[c]
        kidx+=[np.argmax(np.dot(F.T,c))]
```
